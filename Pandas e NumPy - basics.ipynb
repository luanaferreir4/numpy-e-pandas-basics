### .loc, um método muito importante

- Pegar 1 linha
- Pegar linhas de acordo com condições
- Pegar linhas e colunas específicas
- Pegar 1 valor específico

#### .loc funciona de acordo com os *índices*

```python
# pegar uma linha:
tabela_vendas.loc[1]
```

```python
# pegar + de uma linha:
tabela_vendas.loc[1:5] # pegar da linha 1 até a 5.
```

```python
# pegar todas as linhas do Norte Shopping:
# pegar linhas que correspondem a uma condição:
vendas_norteshopping = tabela_vendas.loc[tabela_vendas['ID Loja'] == 'Norte Shopping']
colunas_norteshopping = vendas_norteshopping[['ID Loja', 'Produto', 'Quantidade', 'Valor Unitário', 'Valor Final']]
display(vendas_norteshopping)
display(colunas_norteshopping)
```

```python
# pegar linhas e colunas específicas:
vendas_norteshopping_linha_e_coluna = tabela_vendas.loc[tabela_vendas['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']] # linhas, colunas
display(vendas_norteshopping_linha_e_coluna)
```

```python
# pegar 1 valor específico:
segundo_produto = tabela_vendas.loc[1, 'Produto']
segundo_produto
```

### Como agrupar as informações? Exemplo: faturamento dos produtos.
